```python
#Cleaning Text (RT, Punctuation etc)
#Create new dataframe and new features
list_tw = pd.DataFrame(list_tweet)
list_tw['text'] = list_tw[0]
list_tw['location'] = list_tw[2]

#Removing RT, Punctuation etc
remove_rt = lambda x: re.sub('RT @\w+: '," ",x)
rt = lambda x: re.sub('(@[A-Za-z09]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)'," ",x)
list_tw["text"] = list_tw.text.map(remove_rt).map(rt)
list_tw["text"] = list_tw.text.str.lower()
list_tw.head(10)
```

```python
#Removing Punctuation
def remove_punc(text):
    text="".join([char for char in text if char not in string.punctuation])
    text=re.sub('[0-9]+','',text)
    return text

list_tw['punc']=list_tw['text'].apply(lambda x:remove_punc(x))
```

```python
#Applying Tokenization
def tokenization(text):
    text=re.split('\W+', text)
    return text

list_tw['token']=list_tw['punc'].apply(lambda x:tokenization(x.lower()))
```

```python
#Applying stopwords 
nltk.download('stopwords')
stopword = nltk.corpus.stopwords.words('english')
def remove_stopwords(text):
    text=[word for word in text if word not in stopword]
    return text

list_tw['non_stopword']=list_tw['token'].apply(lambda x:remove_stopwords(x))
```

```python
#Applying Stemmer 
ps = nltk.PorterStemmer()

def stemming(text):
    text=[ps.stem(word) for word in text]
    return text

list_tw['stemmed']=list_tw['non_stopword'].apply(lambda x: stemming(x))
```

```python
#Cleaning Text
def clean_text(text):
    #removing punctuation
    text_lc="".join([word.lower() for word in text if word not in string.punctuation])
    text_rc=re.sub('[0-9]+','', text_lc)
    #tokenization
    token = re.split('\W+', text_rc)
    #removing stopwords and stemming
    text = [ps.stem(word) for word in token if word not in stopword]
    return text
```

```python
#Result (punc, token, stopword, non_stopword, stemmed)
list_tw.head(10)
```

```python
#Applying Countvectorizer
count_vect = CountVectorizer(analyzer=clean_text)
count_vector = count_vect.fit_transform(list_tw['text'])
print('{} Number of reviews {} words'.format(count_vector.shape[0], count_vector.shape[1]))
```

```python
#Function to ngram
def get_top_gram(corpus, ngram_range, n=None):
    vec = CountVectorizer(ngram_range=ngram_range, stop_words='english').fit(corpus)
    bag_of_words = vec.transform(corpus)
    sum_words = bag_of_words.sum(axis=0)
    word_freq = [(word, sum_words[0, idx]) for word, idx in vec.vocabulary_.items()]
    word_freq = sorted(word_freq, key = lambda x: x[1], reverse=True)
    return word_freq[:n]
```

```python
#n1_trigrmas
n1_trigrmas = get_top_gram(list_tw['text'], (3,3), 20)

n1_trigrmas
```
